```python
import numpy as np
import matplotlib.pyplot as plt

# 乱数生成器の設定
#アルゴリズムはPCG64
rng = np.random.default_rng()

# 正規分布に従う乱数を生成（平均0、標準偏差1）
#normal 関数のlocパラメータは平均値（ここでは0）、scaleパラメータは標準偏差（ここでは1）、そしてsizeパラメータで生成する乱数の個数を指定
normal_random_numbers = rng.normal(loc=0, scale=1, size=10000)

# ヒストグラムを描画
plt.hist(normal_random_numbers, bins=300)
plt.xlabel('x')
plt.ylabel('frequency')
plt.xlim(-5, 5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# delimiterをカンマに設定して読み込む
X = np.loadtxt('assignment3.csv', delimiter=',')
shape = X.shape
X_obs = np.zeros((shape[0], shape[1]))  # Use a tuple for the shape
rng = np.random.default_rng()

for i in range (len(X)):
    normal_random_numbers = rng.normal(loc=0, scale=1, size=1)
    X_obs[i, :] = X[i, :] +  normal_random_numbers

np.savetxt('assignment4.csv',X_obs ,delimiter=',')
```
