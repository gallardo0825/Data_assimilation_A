```python
import numpy as np
import matplotlib.pyplot as plt

#F
F = 8.0
#N
N = 40 #40
#dt
dt = 0.01 
#1(N*dt*365)
tend = N*dt*365
nstep = int(tend/dt)

def L96(x):
    dxdt = np.zeros((N))
    for i in range(2, N-1):
        dxdt[i] = (x[i+1]-x[i-2])*x[i-1] - x[i] + F

    dxdt[0] = (x[1]-x[N-2])*x[N-1] - x[0] + F
    dxdt[1] = (x[2]-x[N-1])*x[0] - x[1] + F
    dxdt[N-1] = (x[0]-x[N-3])*x[N-2] - x[N-1] + F

    return dxdt

# L96RK4
def Rk4 (xold, dt):
    k1 = L96(xold)
    k2 = L96(xold + k1*dt/2.)
    k3 = L96(xold + k2*dt/2.)
    k4 = L96(xold + k3*dt)

    #
    xnew = xold + dt/6.0*(k1 + 2.0*k2 + 2.0*k3 + k4)
    
    return xnew

def method():
    xn = np.zeros((N, nstep))#(140)
    t = []

    #(F = 8.0, when initially X20 = F + 0.008 and Xj = F when j != 20 )
    X = F*np.ones(N)
    #j=19
    X[19] = 1.008*F

    for i in range(N):
        xn[i, 0] = X[i]

    for j in range(1, nstep):
        X = Rk4(X, dt)
        xn[:, j] = X[:]
        t.append(dt*j)
    return xn

if __name__ == '__main__':
    xn = method()
    fig = plt.figure(figsize = (8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(xn[0],xn[1],xn[2])
    plt.figure()
    plt.plot(xn[0],xn[1])
```
